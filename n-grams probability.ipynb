```python
# IMPORTING TEXT FILE
with open('bigdata.txt','r') as b:
    txt = b.read()
print(txt)
```

```python
# REMOVE PUNCTUATIONS
import string

txt_np = ''
for i in txt:
    if i not in string.punctuation:
        txt_np += i
print(txt_np)
```

```python
# TOKENIZING
import nltk

tokens = nltk.tokenize.word_tokenize(txt_np)
print(tokens)
```

```python
# CONVERTING TO LOWERCASE
for i in range(len(tokens)):
    tokens[i] = tokens[i].lower()
print(tokens)
```

```python
# FREQUENCY DISTRIBUTION OF UNIGRAMS
unigrams = nltk.ngrams(tokens,1)
unigram_count = nltk.FreqDist(unigrams)
unigram_count = dict(unigram_count)
unigram_count
```

```python
# FREQUENCY DISTRIBUTIONS OF BIGRAMS
bigrams = nltk.ngrams(tokens,2)
bigram_count = nltk.FreqDist(bigrams)
bigram_count = dict(bigram_count)
bigram_count
```

```python
# FREQUENCY DISTRIBUTIONS OF TRIIGRAMS
trigrams = nltk.ngrams(tokens,3)
trigram_count = nltk.FreqDist(trigrams)
trigram_count = dict(trigram_count)
trigram_count
```

```python
n = len(tokens)  # Total no.of works
UNIGRAMS = list(unigram_count.keys())
BIGRAMS = list(bigram_count.keys())
TRIGRAMS = list(trigram_count.keys())
```

```python
# CALCULATING UNIGRAM PROBABILITIES
unigrams_prob = {}
for u in UNIGRAMS:
    unigrams_prob[u] = unigram_count[u]/n
```

```python
unigrams_prob
```

```python
# CALCULATING BIGRAM PROBABILITIES
bigrams_prob = {}
for b in BIGRAMS:
    b_list = list(b)
    onegram = tuple(b_list[:1])
    bigrams_prob[b] = bigram_count[b]/unigram_count[onegram]
bigrams_prob    
```

```python
# CALCULATING TRIGRAM PROBABILITIES
trigram_prob = {}
for t in TRIGRAMS:
    t_list = list(t)
    twogram = tuple(t_list[:2])
    trigram_prob[t] = trigram_count[t]/bigram_count[twogram]
trigram_prob
```

```python
import pandas as pd
pd.set_option('display.max_rows', None)

UNI_GRAM = pd.DataFrame()
UNI_GRAM['UNIGRAM'] = list(unigrams_prob.keys())
UNI_GRAM['PROBABILITY'] = list(unigrams_prob.values())

BI_GRAM = pd.DataFrame()
BI_GRAM['BIGRAM'] = list(bigrams_prob.keys())
BI_GRAM['PROBABILITY'] = list(bigrams_prob.values())

TRI_GRAM = pd.DataFrame()
TRI_GRAM['TRIGRAM'] = list(trigram_prob.keys())
TRI_GRAM['PROBABILITY'] = list(trigram_prob.values())
```
